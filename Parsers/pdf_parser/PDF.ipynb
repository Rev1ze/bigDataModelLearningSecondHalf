```python
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return text
```

```python
# Перебор текста из pdf файла, парсинг
Brak = []
namecompany = []
description = []
rating = []
date = []
field = []
textpub = []
for pdf in all_pdf:
    try:  
        x = extract_text_from_pdf(pdf)
        print(pdf)
        N, R, A, D = pars_pdf(x[0], pdf)
        T = ''.join(x).replace('\n', ' ').replace('\t', ' ').replace('\xa0', ' ').replace('\u200e', ' ').replace('\r', ' ').replace('\uedda', ' ').split('КАК СТАТЬ АВТОРОМ')[1].strip()
        namecompany.append(N)
        rating.append(R)
        date.append(D)
        field.append(A)
        textpub.append(T)
    except Exception as e:
        Brak.append(e)
```

### Получение информаций

```python
result_list = {'title': [], 'description': [], 'rating': [], 'genre': [], 'year': [], 'length': [], 'country' : []} 
```

```python
offset = 0
titles, years, genres, countries, lengths, descriptions,ratings = [],[],[],[],[],[],[]
for j in range(1,9):
    offset += 32
    url = f'https://ntvplus.tv/movies/series/ajax/series?token=&sort=&offset={offset}'
    page = requests.get(url)
    soup = bs(page.text,'html.parser')
    titles_links = soup.findAll('a', 'poster--title')    
    for i in titles_links:
        
        this_film = i["href"]
        if this_film == "/watch/22928-i-prosto-tak-dokumentalnyj-film": 
            
            continue
        url_about = "https://ntvplus.tv"
        try:
            page = requests.get(url_about + i["href"])
            soup_second = bs(page.text, 'html.parser')
            genre = soup_second.find_all('div', "show-info--field-r")[1].span.text
            year = soup_second.find_all('div', "show-info--field-r")[2].a.text
            country = soup_second.find_all('div', "show-info--field-r")[3].a.text
            length = soup_second.find_all('div', "show-info--field-r")[5].text 
            description = soup_second.find_all('div', "show-info--field-r")[8].text
            int(year)
            
            

            try:
                rating = soup_second.find('div','show-info--counter').span.text
                rating = float(rating)
                print(f"((((((((((())))))))))) {rating} _))))))))")
            except:
                print(f"__________________- {rating} ___________-")

                rating = 0
        except Exception as e:
            print("qweqweqweqweqwe")
            print(e)
            continue
        titles.append(i.text)
        genres.append(genre)
        ratings.append(rating)
        years.append(year)
        countries.append(country)
        lengths.append(length)
        descriptions.append(description)


        print(f"{i.text}\n{rating}\n{year}\n{country}\n{length}\n{description}\n{genre}\n___________________________")

    
```

```python
result_list["title"] = titles
result_list["genre"] = genres
result_list["country"] = countries
result_list["length"] = lengths
result_list["description"] = descriptions
result_list["year"] = years
result_list["rating"] = rating

```
